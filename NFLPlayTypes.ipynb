```python
import pandas as pd
import glob
```

```python
cols = ['Date', 'desc', 'PlayType']
pxp = []
for year in range(2009, 2017):
    files = glob.glob('NFL_Data/pxp_{}.csv'.format(year))
    for file in files:
        current = pd.read_csv(file, usecols=cols)
        pxp.append(current)
        
pxp = pd.concat(pxp)
```
