```python
import pandas as pd
import glob
```

```python
pxp = []
cols = ['year', 'description', 'type']

for year in range(2001, 2018):
    files = glob.glob('NCAA-Football-Data/{}PXP/*.csv'.format(year))
    for file in files:
        plays = pd.read_csv(file, usecols=cols)
        pxp.append(plays)
        
pxp = pd.concat(pxp)
```

```python
types = pd.unique(pxp['type'])
types
```

```python
for playType in types:
    print(str(playType) + ':')
    for year in range(2001, 2018):
        if len(pxp.loc[(pxp['type'] == playType) & (pxp['year'] == year)]) != 0:
            item = pxp.loc[(pxp['type'] == playType) & (pxp['year'] == year)].sample(1).iloc[0]
```
