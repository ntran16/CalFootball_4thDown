```python
import pandas as pd
import glob
```

```python
pxp = []
cols = ['year', 'week', 'description', 'type', 'down']

for year in range(2001, 2018):
    files = glob.glob('NCAA-Football-Data/{}PXP/*.csv'.format(year))
    for file in files:
        plays = pd.read_csv(file, usecols=cols)
        pxp.append(plays)
        
pxp = pd.concat(pxp)
```

```python
pxp
```

```python
types = pd.unique(pxp['type'])
types
```

```python
for playType in types:
    print(str(playType) + ':')
    for year in range(2001, 2018):
        if type(playType) == str:
            items = pxp.loc[(pxp['type'] == playType) & (pxp['year'] == year)]
        else:
            items = pxp.loc[(type(playType) != str) & (pxp['year'] == year)]
        if len(items) != 0:
            item = items.sample(1).iloc[0]
            print('Year', str(item['year']), 'Week', str(item['week']) + ',', item['description'])
    print()
```
